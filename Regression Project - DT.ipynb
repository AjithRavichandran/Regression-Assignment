```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_details.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor as DTR 
regressor=DTR(criterion='squared_error',splitter='random',max_features=None)
regressor.fit(input_train,output_train)
```

```python
import matplotlib.pyplot as plt 
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
predict=regressor.predict(input_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(output_test,predict)
```
